```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies = movies.merge(credits, on="title")
```

```python
movies = movies[["movie_id", "title", "overview", "genres", "keywords", "cast", "crew"]]
movies.head(1)
```

```python
movies.isnull().sum()
movies.dropna(inplace=True)
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

def director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
          L.append(i['name'])
          break
    return L

def convert2(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter!=3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movies['keywords']= movies['keywords'].apply(convert)
movies['genres']=movies['genres'].apply(convert)
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']= movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']= movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']= movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']= movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
movies['tag'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tag']]
new_df.head(1)
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:" ".join(x))
new_df.head(1)
```

```python
new_df['tag'][0]
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tag'] = new_df['tag'].apply(stem)
```

```python
new_df['tag'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
import pickle
pickle.dump(new_df, open('movies.pkl', 'wb'))
pickle.dump(similarity,open('similarity.pkl','wb'))
```

```python
recommend('The Dark Knight Rises')
```
